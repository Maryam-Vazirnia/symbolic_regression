# <p style="direction:rtl;text-align:right;"> مقدمه:
   
<p style="direction:rtl;text-align:right;">ایده اصلی ژنتیک الگوریتم مبنی بر نظریه تکاملی داروین است یعنی ان دسته از صفات طبیعی که با طبیعت سازگاری بیشتری دارند شانس بقای بیشتری دارند. و افراد یک حامعه از طریق جفت گیری نسل جدیدی ایجاد می کنند که در نسل جدید اغلب افراد سازگاری بیشتری با طبیعت دارند. و در هر نسل شانس انتقال خصوصیات گونه های بهتر به نسل بعد بیشتر است. درنتیجه با گذشت زمان و در طول نسل ها گونه ها تکامل می یابند. این فرآیند یک فرآیند تصادفی و همگراست. و الگوریتم ژنتیک درواقع شبیه سازی این فرآیند تکاملی است. جواب این برنامه در قالب یک بردار یا ماتریس  است که به عنوان کروموزم شناخته می شود. و مجموعه ای از کوروموزم ها در هر نسل را جمعیت می نامیم. میزان سازگاری کوروموزم ها با محیط نیز توسط تابع فیتنس ارزیابی می شود.به عبارت دیگه میزان مقبولیت یک جواب یا همان کوروموزم براساس این تابع محاسبه می شود. که تابع فیتنس با توجه به برنامه مشخص می شود. و در هر نسل کوروموزم های برتر شانس بیشتری برای تولید مثل دارند. فرایند تولید مثل نیز با استفاده از سه عملگر ژنتیکی تقاطع جهش و کپی انجام می شود.
    
<p style="direction:rtl;text-align:right;">"سیمبولیک رگرسین" از روش های برنامه نویسی ژنتیک است که برای تحلیل دیتاهای مشاهداتی؛ عباراتی ریاضی می سازد.یا به عبارت دیگه ساختن عبارات ریاضی ای که بهترین توصیف رو از دیتاها داشته باشه. و میتونیم بگیم **برنامه نویسی ژنتیک یک فرآیند بهینه سازی تصادفی است.**

این پروسه از یک فضای عبارت های ریاضی بهره میگیرد که از دو چیز استفاده می کند یکی عملگرها یکی توابع که توسط کاربر برای برنامه تعریف می شود. به طور مثلا چهار عمل اصلی به عنوان عملگر و توابعی مثل سینوس لگاریتم و ... به عنوان توابع برای برنامه تعریف می شود. دقت خوب این برنامه و مستقل از مدل بودن و در نتیجه نداشتن بایاس از مزیت های آن است اما خروجی توابع پیچیده و عدم نمایش مستقیم خطاها از عیوب این روش است.

### <p style="direction:rtl;text-align:right;">مشکلات: 
گفتیم اندازه گیری فیتنس برنامه از طریق برخی عملکردها است که پیش بینی های برنامه را در مقایسه با حقیقت اساسی ارزیابی می کند. اما با توابعی مانند تقسیم در function_set ، به مشکل برمیخوریم. مثلا اگر مخرج شما صفر یا نزدیک به صفر باشد؟بنابراین برای کل مجموعه خطاخواهیم داشت! بنابراین از برخی استدلال ها استفاده می کنیم. تا عملکردهایی مانند تقسیم بتوانند هر ورودی را بپذیرند و همچنان یک مقدار معتبر برگردانند. از جمله:
    
• division, if the denominator lies between -0.001 and 0.001, returns 1.0.

• square root returns the square root of the absolute value of the argument.

• log returns the logarithm of the absolute value of the argument, or for very small values less than 0.001, it returns 0.0.

• inverse, if the argument lies between -0.001 and 0.001, returns 0.0.

<p style="direction:rtl;text-align:right;">به این ترتیب ،یک خروجی عددی معتبر را می توان تضمین کرد.عملکرد function.make_function () نیز در مقابل عملیات منفی یا نزدیک به صفر را میتوانیم برایش استدلال هایی تعیین کنیم.

<p style="direction:rtl;text-align:right;">یکی دیگه از چیزهایی که برای اجرای موفق ژنتیک پروگرام لازم است
توجه به برخی نکات کوچک است. مثلا دامنه ی مقادیر ثابت را روی مقدار مناسبی تنظیم کنید تا زودتر به هدف برسید. مثلا اگر مقدار ثابت شما حدودا ۵۰۰ باشد و دامنه مقادیر شما در برنامه از صفر تا ۱۰ باشد این باعث ایجاد خطا و همچنین سپری شدن زمان زیادی در اجرای برنامه می شود. پارامتر const_range رنج ثوابت را کنترل می کند که بصورت پیش فرض بین -۱ تا ۱ است.  یا اگر از توابع مثلثاتی استفاده می کنید تمام زوایا به صورت رادیان اندازه گیری می شوند.(و درنظر بگیرید که ورودی ها تاثیر تناوبی روی هدفتان دارند یا نه). یا مثلا اگر فکر میکند که حل این برنامه یک فرمول بزرگ است یا متغیرهای زیادی دارید پس عمق بیشتری برای درختتان تعیین کنید. 
    

<p style="direction:rtl;text-align:right;">جمعیت نسل اول به طور تصادفی از توابع و متغیرهای موجود انتخاب می شود و اعتبار ضعیفی هم دارند اما می شود کاربر جمعیت اولیه را خود ارایه دهد. (؟؟)
از پارامترهای دیگر که در کد باید به آن توجه کنیم
init_depth
یعنی عمق برنامه در اولین نسل است. که یک تاپل از دو عدد صحیح است و دامنه عمق اولیه ای را که نسل اول می تواند داشته باشد مشخص می کند.این محدوده فقط مربوط به نسل اول می شود.(فقط نسل اول؟؟) و دامنه پیش فرض 2 - 6 است که به طور کلی نقطه شروع خوبی است.

<p style="direction:rtl;text-align:right;">نکته بعد سایز جمعیت است. اگر متغیرهای بسیار کمی داشته باشیم و یا مجموعه عملکردهای محدودی داشته باشیم جمعیت کم هم برایمان کفایت می کند. اما اگر تعداد متغیرهایمان زیاد باشد یا به طور پیش فرض انتظار یک برنامه ی بزرگ را داشته باشیم باید سایز جمعیت را بزرگ انتخاب کنیم.و نیز ،باید پارامتر 
init_method
را به طور مناسبی برای داده هایمان قرار دهیم. که میتواند انتخاب از بین موارد زیر باشد:
'grow', 'full', or 'half and half'

<p style="direction:rtl;text-align:right;"> I. grow: در همه موارد، گره ریشه باید یک تابع باشد. (بر خلاف یک متغیر یا ثابت که لازم نیست تابع باشند).
در این روش، گره ها به صورت تصادفی هم توابع و هم از پایانه ها انتخاب می شوند و درختان شروع به رشد می کنند و عمق کمی هم دارند. در این روش درختان نامتقارن رشد می کنند. زیرا قبل از رسیدن به  عمق حداکثری احتمال اینکه یک گره ترمینال انتخاب شود زیاد است. نکته این است که اگر مجموعه داده های شما متغیرهای زیادی داشته باشد، به برنامه های زیاد با عمق کم منجر می شود و اگر متغیرهای بسیار کمی دارید و یک مجموعه عملکرد بزرگ را انتخاب کرده اید، به احتمال زیاد برنامه هایی را مشاهده می کنید که به حداکثر عمق مشخص شده توسط شاخص عمق نزدیک می شوند.
    
<p style="direction:rtl;text-align:right;"> II. full: در این روش گره ها تنها از مجموعه فانکشن-ست انتخاب می شوند تا رسیدن به عمق حداکثر، و سپس پایانه ها انتخاب می شوند. این روش تمایل به رشد درختان "بوته ای" متقارن دارد. 
اما پیش فرض روش "half and half" است. به اینگونه است که برنامه به صورت مخلوط رشد می کند یعنی نیمی از جمعیت متد-اولیه را روی "full" تنظیم می شوند و نیمی دیگر "grow" می کنند.

<p style="direction:rtl;text-align:right;">حالا که جمعیت زیادی از برنامه ها را داریم، باید تصمیم بگیریم که کدام یک از آنها به نسل بعدی تبدیل شوند. این کار را تورنمنت انجام میدهد. به این صورت که از میان جمعیت، یک زیر مجموعه کوچکتر به صورت تصادفی برای رقابت انتخاب می شود که اندازه آن توسط پارامتر tournament_size
کنترل می شود (تعداد برنامه هایی که رقابت می کنند با هم برای رفتن به نسل بعدی). مناسب ترین فرد 
(fittest individual)

### <p style="direction:rtl;text-align:right;"> Bloat :
 
<p style="direction:rtl;text-align:right;"> اندازه برنامه می تواند به دو روش اندازه گیری شود: عمق و طول آن. عمق برنامه یعنی فاصله از گره ریشه تا دورترین گره برگ است. مثلا یک برنامه با یک مقدار واحد (یعنیy = X0) دارای عمق صفر است. طول یک برنامه هم تعداد عناصر موجود در فرمول است که با تعداد کل گره ها برابر است.

# <p style="direction:rtl;text-align:right;">  مثال اول

<p style="direction:rtl;text-align:right;"> در این مثال دو متغیر وابسته طبق زیر داریم:
$$y=x_0^2-x_1^2+x_1-1$$

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x0 = np.arange(-1, 1, 1/10.)
x1 = np.arange(-1, 1, 1/10.)
x0, x1 = np.meshgrid(x0, x1)
y_truth = x0**2 - x1**2 + x1 - 1

# print(x0)
# print(y_truth)
# print(len(x0))
# print(np.shape(x1))
# print(np.shape(y_truth))

ax = plt.figure().gca(projection='3d')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
surf = ax.plot_surface(x0, x1, y_truth, rstride=1, cstride=1,color='red', alpha=0.5)
plt.show()
```

```python
#  We can create some random training and test data that lies on this surface too:
from sklearn.utils import check_random_state
rng = check_random_state(0)
# Training samples
X_train = rng.uniform(-1, 1, 100).reshape(50, 2)
y_train = X_train[:, 0]**2 - X_train[:, 1]**2 + X_train[:, 1] - 1
# Testing samples
X_test = rng.uniform(-1, 1, 100).reshape(50, 2)
y_test = X_test[:, 0]**2 - X_test[:, 1]**2 + X_test[:, 1] - 1
# print(X_train)
# print(y_train)
# print(X_test)
# print(y_test)
```

```python
import gplearn as gpl        
from gplearn.genetic import SymbolicRegressor
est_gp = SymbolicRegressor(population_size=5000,generations=20, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,p_hoist_mutation=0.05,
                           p_point_mutation=0.1,max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=0)
est_gp.fit(X_train, y_train)
```

```python
print(est_gp._program)
# sub(add(-0.999, X1), mul(sub(X1, X0), add(X0, X1)))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

est_tree = DecisionTreeRegressor()
est_tree.fit(X_train, y_train)
est_rf = RandomForestRegressor()
est_rf.fit(X_train, y_train)
```

```python
y_gp = est_gp.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
score_gp = est_gp.score(X_test, y_test)
y_tree = est_tree.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
score_tree = est_tree.score(X_test, y_test)
y_rf = est_rf.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
score_rf = est_rf.score(X_test, y_test)
fig = plt.figure(figsize=(12, 10))
for i, (y, score, title) in enumerate([(y_truth,None, "Ground Truth"),
                                       (y_gp, score_gp, "SymbolicRegressor"),
                                       (y_tree, score_tree, "DecisionTreeRegressor"),
                                       (y_rf, score_rf, "RandomForestRegressor")]):
    ax = fig.add_subplot(2, 2, i+1, projection='3d')
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    surf = ax.plot_surface(x0, x1, y, rstride=1, cstride=1, color='green', alpha=0.5)
    points = ax.scatter(X_train[:, 0], X_train[:, 1], y_train)
    if score is not None:
        score = ax.text(-.7, 1, .2, "$R^2 =\/%.6f$" % score, 'x', fontsize=14)
    plt.title(title)
plt.show()
```

```python
##  symbolic regression
est_gp=SymbolicRegressor(population_size=500,generations=7,
                        metric='rmse',stopping_criteria=0.01,
                        function_set=('add','sub','mul','sin','log'),
                        p_crossover=0.7,p_subtree_mutation=0.1,p_hoist_mutation=0.05,
                        p_point_mutation=0.1,max_samples=0.9,verbose=1,
                        parsimony_coefficient=0.01,random_state=0,low_memory=True)
```

```python
#  فیت کردن ایکس و وای

est_gp.fit(X_train,y_train)
```

```python
# به دست آوردن نتیجه ی نهایی یا همان فرمول مرتبط کردن ایکس و وای که به شکل درخته

print(est_gp._program)
#  add(add(sin(X0), sin(X0)), add(sub(add(sin(X0), sin(add(X0, 0.004))), log(X0)), sin(X0)))
```

```python
X_test = np.linspace(0.0,20,100).reshape(-1,1)
# X_test=rng.uniform(1,20,100).reshape(-1,1)
y_predict = est_gp.predict(X_test)
# y_predict
```
